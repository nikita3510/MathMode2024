```python
Посчитайте две матрицы: B=A^TA  и C=AA^T . Что вы можете о них сказать? Какого они размера? В чём их сходство и отличие? 
Найдите сумму диагональных элементов матриц  B и C. Посчитайте детерминант и ранг матриц A ,B  и C, посмотрите на спектр 
(набор собственных значений) матриц B и C, какие выводы вы можете сделать? 
(подсказка: используйте готовые функции из библиотеки np.linalg).
```

```python
# Создаем матрицу A размером 3x4
A = np.random.rand(4, 4)

# Вычисляем матрицы B = A^T * A и C = A * A^T
B = np.dot(A.T, A)
C = np.dot(A, A.T)

# Выводим информацию о матрицах B и C
print("Матрица B:")
print(B)
print("Размер матрицы B:", B.shape)

print("\nМатрица C:")
print(C)
print("Размер матрицы C:", C.shape)

# Находим сумму диагональных элементов матриц B и C
sum_diag_B = np.trace(B)
sum_diag_C = np.trace(C)
print("\nСумма диагональных элементов матрицы B:", sum_diag_B)
print("Сумма диагональных элементов матрицы C:", sum_diag_C)

# Находим детерминант и ранг матриц A, B и C
det_A = np.linalg.det(A)
rank_A = np.linalg.matrix_rank(A)

det_B = np.linalg.det(B)
rank_B = np.linalg.matrix_rank(B)

det_C = np.linalg.det(C)
rank_C = np.linalg.matrix_rank(C)

print("\nДетерминант и ранг матрицы A:")
print("Детерминант:", det_A)
print("Ранг:", rank_A)

print("\nДетерминант и ранг матрицы B:")
print("Детерминант:", det_B)
print("Ранг:", rank_B)

print("\nДетерминант и ранг матрицы C:")
print("Детерминант:", det_C)
print("Ранг:", rank_C)

# Находим собственные значения матриц B и C
eigenvalues_B = np.linalg.eigvals(B)
eigenvalues_C = np.linalg.eigvals(C)

print("\nСпектр (собственные значения) матрицы B:")
print(eigenvalues_B)

print("\nСпектр (собственные значения) матрицы C:")
print(eigenvalues_C)

```
