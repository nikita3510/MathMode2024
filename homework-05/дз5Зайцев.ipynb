```python
from PIL import Image
import numpy as np
from numpy.linalg import svd
import os

img = Image.open('picture.jpg')
img_array = np.array(img)
N, I, C = np.linalg.svd(img_array, full_matrices=False)
k = 100
g = np.diag(I[:k])
a =  np.dot(g, C[:k, :])
aT = a.T
f = np.dot(N[:, :k], aT)
f = Image.fromarray(np.asarray(img_array, dtype=np.uint8)).save(f'{img}.png')
T, U, Sh = svd(img_array, full_matrices=False)
col = [1, 2, 10, 30, 100]
PL = []
for k in col:
    carray = np.dot(T[:, :k] * U[:k], Sh[:k, :])
    PLG = Image.fromarray(np.asarray(img_array, dtype=np.uint8))
    PL.append(PLG)
    PLG.save(f'compressed_{k}.png')
orig = os.path.getsize('picture.jpg')
for k, img in zip(col, PL):
    com = os.path.getsize(f'compressed_{k}.png')
    print("The number of singular values: ",k)
    print("Compressed image size:",com,"bytes")
print("The size of the original image: ",{orig},"bytes")
```

Этот код выполняет сжатие изображения с использованием сингулярного разложения (SVD).

1. Сначала изображение загружается с помощью библиотеки PIL и преобразуется в массив NumPy.
2. Затем применяется SVD к массиву изображения, чтобы получить матрицы N, I, C. Выбирается определенное количество сингулярных значений (k) для сжатия.
3. Создается диагональная матрица g из первых k сингулярных значений, которая перемножается с матрицей C, а затем с транспонированной матрицей a.
4. Вычисляется новое сжатое изображение f путем умножения матриц N и aT.
5. Сжатое изображение сохраняется в формате PNG.
6. Затем выполняется SVD для оригинального изображения и создаются несколько вариантов сжатых изображений с разным количеством сингулярных значений (col).
7. Сжатые изображения также сохраняются в формате PNG.
8. Далее происходит подсчет размера оригинального изображения и каждого сжатого изображения в байтах.
9. Размеры каждого сжатого изображения выводятся на экран.
